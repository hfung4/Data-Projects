## Substansive effects of Management Practices on Motivation
- The purpose of this notebook is compute and produce data visualizations of the substansive effects of management practices on motivation (measured by WPI, Work Preference Inventory).

- Substansive effect/significance basically refers to whether an observed efect is large enough to be meaningful.  It is an additional dimension to look at regression results (aside from statisitical significance).  In other words, a regressor can be statistical signficant, but its effect size can be too small to be substansive. 

```python
'''data'''
import pandas as pd
import numpy as np
from scipy import stats
import statistics

'''plots'''
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Autonomy (P3)**

```python
p3 = pd.read_csv("../data/p3/data_p3.csv")
```

```python
p3_imot= p3.intrinsic_motivation
```

```python
p3_imot.head()
```

```python
p3_imot.dropna(inplace=True)
```

```python
# compute the "target"
median = statistics.median(p3_imot)
mean = statistics.mean(p3_imot)
effect_size = 0.319
target = median + effect_size
```

```python
print(mean,median,target)
```

```python
fig = plt.figure(figsize=(12,5)) 
sns.distplot(p3_imot,kde=False ,color='k');
min_ylim, max_ylim = plt.ylim()
plt.text(p3_imot.median()*1.005, max_ylim*0.9, 'Median: {:.2f}'.format(p3_imot.median()))
plt.axvline(p3_imot.median(), color='k', linestyle='dashed', linewidth=1)

plt.text(p3_imot.mean()*1.005, max_ylim*0.2, 'Mean: {:.2f}'.format(p3_imot.mean()))
plt.axvline(p3_imot.mean(), color='k', linestyle='dashed', linewidth=1)

plt.axvline(x=4.54, color='k', linewidth=1)
plt.text(4.54*1.005, max_ylim*0.5, 'Increase to: {:.2f}'.format(4.54))

plt.ylabel('Frequency', fontsize=12)
plt.xlabel('Intrinsic motivation', fontsize=12)

plt.arrow(p3_imot.median(), 600000, 4.54-p3_imot.median()-0.04,0, head_width=20000, head_length=0.05, fc='k', ec='k')

plt.savefig('output/effect_size_autonomy_P3.png', dpi=300, bbox_inches='tight')
```

```python
percentile=stats.percentileofscore(p3_imot, target, 'mean')
```

```python
print("It is estimated that a unit change in the autonomy score will result in a change from the median intrinsic motivation score of {} to {}. This corresponds to an increase from the 50th intrinsic motivation score to the {}th percentile intrinsic motivation score.".format(round(median,4),round(target,4), round(percentile,4)))
```

**resources (H4,P2), use mean**

```python
X_mean = p1_p2.resources.mean()
X_mean
```

```python
# Y = Bo + B1*X1 + B2*nonver + B3*X1*nonver +u
median = statistics.median(p2_imot)
mean = statistics.mean(p2_imot)
effect_size = -0.404+0.149*X_mean   # B2 + B3X_bar     (NOTE: (X_bar,Y_bar) lies on regression line)
print(effect_size)
# For a non-verifable agency, an unit increase in qual_supervisor will result in change from the median imot score
target = mean + effect_size 
```

```python
print("It is estimated from the mean adequate resources and intrinsic motivation score, a change from verifiable to non-verifiable tasks will result in a change in intrinsic motivation from its mean {} to {}. This corresponds to an increase from the {}th percentile intrinsic motivation score to the {}th percentile intrinsic motivation score.".format(round(mean,4),round(target,4), round(stats.percentileofscore(p2_imot, mean, 'mean'),4), round(stats.percentileofscore(p2_imot, target, 'mean'),4)))
```
